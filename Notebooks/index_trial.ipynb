```python
# Define a list of documents to be indexed
documents = [
    {"id": 1, "text": "This is the first document."},
    {"id": 2, "text": "This is the second document."},
    {"id": 3, "text": "This is the third document."}
]

index_name = 'debateorg-preprocessed'

# Create the index with dynamic mapping
es.indices.create(index=index_name, body={
    "mappings": {
        "dynamic": True
    }
}, ignore=[400], request_timeout=30)

# Index the documents
print("Start indexing")
for i, doc in enumerate(documents):
    if not i % 10:
        print("Documents indexed: ", i)
    response = es.index(
        index=index_name,
        id=doc['id'],
        document=doc
    )
    print("Indexed document: ", response)

print("Indexing complete")
```
